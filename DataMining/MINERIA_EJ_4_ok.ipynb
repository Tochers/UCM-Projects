## Clustering Jer√°rquico
**Prueba con los 4 tipos de Linkage y observaci√≥n de posible n√∫mero de clusters**

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from time import time


def plot_dendogram(model, **kwargs):
  
  '''
  Esta funci√≥n extrae la informaci√≥n de un modelo AgglomerativeClustering
  y representa su dendograma con la funci√≥n dendogram de scipy.cluster.hierarchy
  '''
      
  counts = np.zeros(model.children_.shape[0])
  n_samples = len(model.labels_)
  for i, merge in enumerate(model.children_):
    current_count = 0
    for child_idx in merge:
      if child_idx < n_samples:
        current_count += 1  # leaf node
      else:
        current_count += counts[child_idx - n_samples]
    counts[i] = current_count
  
  linkage_matrix = np.column_stack([model.children_, model.distances_,
                                        counts]).astype(float)
  
  # Plot
  dendrogram(linkage_matrix, **kwargs)
  plt.show()
```

```python
for linkage in ("ward", "average", "complete", "single"):
    clustering = AgglomerativeClustering(linkage=linkage, n_clusters=None,
                            distance_threshold = 0)
    t0 = time()
    clustering.fit(X_scaled)
    print("%s :\t%.2fs" % (linkage, time() - t0))
    
    plt.clf()
    plot_dendogram(clustering)
```

Por el clustering jer√°rquico, parecen definirse entre 3 y 5 clusters. Aqu√≠ elijo tipo de *linkage ward* que define de forma mas precisa 3 grupos.

## Clustering por K-Means

```python
from scipy.spatial.distance import cdist, pdist

def scree_plot_kmeans(data,n_max):
  range_n_clusters = range(2, n_max)
  X_scaled = scale(data)
  inertias = []
  silhouette = []
  var_perc = []
  
  for n_clusters in range_n_clusters:
      modelo_kmeans = KMeans(
                          n_clusters   = n_clusters, 
                          n_init       = 20, 
                          random_state = 123
                      )
      modelo_kmeans.fit(X_scaled)
      cluster_labels = modelo_kmeans.fit_predict(X_scaled)
      inertias.append(modelo_kmeans.inertia_)
      silhouette.append(silhouette_score(X_scaled, cluster_labels))
      tss = sum(pdist(X_scaled)**2)/X_scaled.shape[0]
      bss = tss - modelo_kmeans.inertia_
      var_perc.append(bss/tss*100)
      
  fig, ax = plt.subplots(1, 3, figsize=(16, 6))
  ax[0].plot(range_n_clusters, inertias, marker='o')
  ax[0].set_title("Scree plot Varianza intra")
  ax[0].set_xlabel('N√∫mero clusters')
  ax[0].set_ylabel('Intra-cluster (inertia)')
  
  ax[1].plot(range_n_clusters, silhouette, marker='o')
  ax[1].set_title("Scree plot silhouette")
  ax[1].set_xlabel('N√∫mero clusters')
  ax[1].set_ylabel('Media √≠ndices silhouette');
  
  ax[2].plot(range_n_clusters, var_perc, marker='o')
  ax[2].set_title("Scree plot % Varianza")
  ax[2].set_xlabel('N√∫mero clusters')
  ax[2].set_ylabel('% de varianza explicada')
```

```python
scree_plot_kmeans(df_grouped,10)
plt.show()
```

En base a las 10 variables escogidas en forma subjetiva no hay un claro patr√≥n que nos permita escoger claramente el n¬∞ de clusters utilizando el indice intra-cluster. A priori podr√≠amos decir que en base al clustering jer√°rquico podr√≠amos definir 3 clusters y la silueta del k-means parece coincidir, no obstante el gr√°fico de la derecha nos indica que 3 clusters solo nos explicar√≠an un poco m√°s del 50% de la varianza.

**Prueba con MinMaxScaler para evaluar cambios en n¬∞ de clusters.**

```python
km_scores= []
km_silhouette = []
vmeasure_score =[]
db_score = []
for i in range(2,10):
    km = KMeans(n_clusters=i, random_state=0).fit(X_scaled)
    preds = km.predict(X_scaled)
    
    print("Score for number of cluster(s) {}: {}".format(i,km.score(X_scaled)))
    km_scores.append(-km.score(X_scaled))
    
    silhouette = silhouette_score(X_scaled,preds)
    km_silhouette.append(silhouette)
    print("Silhouette score for number of cluster(s) {}: {}".format(i,silhouette))
    
    db = davies_bouldin_score(X_scaled,preds)
    db_score.append(db)
    print("Davies Bouldin score for number of cluster(s) {}: {}".format(i,db))
```
