```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(
    r'datasets/Info_pais.csv',
    encoding='ISO-8859-1',
    delimiter=';',
    decimal=','
)
```

```python
df.head()
```

```python
df_order = df.sort_values(
    'Poblacion',
    ascending=False
)
```

```python
df_order.head(15)
```

```python
plt.scatter(df_order['Renta per capita'],df_order['Esperanza de vida']);
```

```python
plt.scatter(df_order['Renta per capita'],df_order['Esperanza de vida'])

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida');
```

```python
df_order.head()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / max(df_order['Poblacion'])
```

```python
df_order.head()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / (max(df_order['Poblacion']) / 10000)
```

```python
df_order.head()
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita Vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida');
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita Vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita Vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita Vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i],df_order['Esperanza de vida'][i])
    )
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápita Vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);

df_order.reset_index(drop=True, inplace=True);

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i],df_order['Esperanza de vida'][i])
    )
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'],
    alpha=0.5
)

plt.title('Renta per cápita Vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10);

df_order.reset_index(drop=True, inplace=True);

for i in range(0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i],df_order['Esperanza de vida'][i])
    )
```

## Conclusiones

EL objetivo del presente análisis gira en torno a la relación existente entre las variables **Renta per cápita** (medida en dólares estadounidenses al año) y la **Esperanza de vida** (medida en años de vida), para conocer si aquellos países con una mayor renta per cápita tienen una mayor expectativa de vida.

A primera vista podemos observar que los países con la mayor Renta per cápita tienen una esperanza de vida superior a la mundial para 2020 (73 años. <i>Fuente: World Bank Data</i>), lo que llevaría a suponer que una relación directa entre ambas variables. Sin embargo, a raíz de la información y los gráficos obtenidos observamos que dentro de los países con menor Renta per cápita en el mundo, la esperanza de vida oscila de forma amplia, con países que no superan los 60 años y otros que superan los 80 años de expectativa de vida dentro de una misma franja de RPC.

Por este motivo, podemos concluir que la RPC no explica las variaciones en la esperanza de vida, sino que deberían considerarse variables adicionales al análisis para descubrir por qué países con una RPC baja pueden lograr una esperanza de vida igual o superior a países con una RPC muy elevada.
